<p></p>
<p>&nbsp;</p></div>

```
# mengambil hanya data untuk produk 'A'
produk_A = raw_data[raw_data['Produk'] =='A']
 
# menghitung rerata pendapatan menggunakan method .mean pada objek pandas DataFrame
print (produk_A['Pendapatan'].mean())
 
# menghitung rerata pendapatan menggunakan method .mean pada objek pandas DataFrame dengan numpy
print (np.mean(produk_A['Pendapatan']))
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Median</h2><p>Median adalah salah satu ukuran pusat yang nilainya terletak di tengah titik data. Sebagai gambaran, jika kita memiliki titik data bernilai 1, 2, 3, 4, 4, 5, 6 maka median dari sekumpulan titik data tersebut adalah 4. Namun jika kita memiliki titik data bernilai 1, 2, 3, 3 maka media dari sekumpulan titik data tersebut adalah:</p>
<p></p>
<p>Kita dapat menemukan nilai median dengan menggunakan method&nbsp;<code>.median()</code>&nbsp;pada numpy maupun pandas sebagaimana contoh berikut:</p>
<p></p></div>

```
print (raw_data)
# Hitung median dari usia menggunakan pandas
print (produk_A['Pendapatan'].median())

# Hitung median dari pendapatan menggunakan numpy
print (np.median(produk_A['Pendapatan']))
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Modus</h2><p>Modus didefinisikan sebagai data yang memiliki frekuensi kemunculan terbanyak/terbesar. Sebagai contoh, jika terdapat titik data seperti berikut: 1, 1, 1, 1, 2, 3, 3, 4 maka modus dari data tersebut adalah 1 karena 1 muncul sebanyak 4 kali, lebih banyak dibanding titik data lainnya.</p>
<p>Kita dapat menggunakan&nbsp;<em>method</em>&nbsp;<code>.count_values()</code>&nbsp;pada pandas sebagaimana contoh berikut:</p>
<p></p></div>

```
# Melihat jumlah dari masing-masing produk
print(raw_data['Produk'].value_counts())
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Kuantil</h2><p>Kuantil adalah nilai-nilai data yang membagi data yang telah diurutkan sebelumnya menjadi beberapa bagian yang sama besar ukurannya. Beberapa ukuran fraktil ini diantaranya adalah:</p>
<ul>
<li>Kuartil: Adalah kuantil yang membagi data menjadi empat bagian sama besar. Kuartil ke-2 dari adalah median dari data tersebut.</li>
<li>Desil: Adalah kuantil yang membagi data menjadi 10 bagian sama besar.</li>
<li>Persentil: Adalah kuantil yang membagi data menjadi 100 bagian sama besar.</li>
</ul>
<p>Untuk mencari fraktil dari data, kita dapat menggunakan&nbsp;<em>method</em>&nbsp;<code>.quantile</code>&nbsp;dari pandas atau numpy sebagaimana contoh berikut:</p>
<p></p></div>

```
# mencari median atau 50% dari data menggunakan pandas
print(raw_data['Pendapatan'].quantile(q=0.5))
 
# mencari median atau 50% dari data menggunakan pandas
print(np.quantile(raw_data['Pendapatan'],q=0.5))
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Agregasi Data dengan method .agg()</h2><p>Ada kalanya kita ingin menghitung sekaligus beberapa ukuran, misalnya menghitung nilai mean sekaligus menghitung nilai median. Kita dapat melakukan kedua hal tersebut dengan menggunakan method&nbsp;<code>.agg()</code>&nbsp;pada objek pandas DataFrame sebagaimana contoh berikut:</p>
<p></p></div>

```
# menghitung rerata dan median 'Pendapatan' dan 'Harga'
print(raw_data[['Pendapatan','Harga']].agg([np.mean, np.median]))
 
# menghitung rerata dan median Pendapatan dan Harga dari tiap produk
print(raw_data[['Pendapatan','Harga','Produk']].groupby('Produk').agg([np.mean, np.median]))
```

Untuk mencari kuartil ke-1 dan ke-3 dari kolom 'Pendapatan' maka kode yang tepat adalah:

raw_data['Pendapatan'].quantile([0.25, 0.75])


Jika kita ingin menghitung rata-rata dan median dari variabel 'Pendapatan' dan 'Harga', maka kita dapat menggunakan method:

np.mean(x) dan np.quantile(x, q=0.5)

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Proporsi Kategori</h2><p>Proporsi kategori adalah ukuran sebaran yang paling sederhana dari ukuran sebaran pada data nomisal dan ordinal. Secara matematis dapat dirumuskan sebagai:</p>
<p></p>
<p></p></div>

```
# cari proporsi tiap Produk
print (raw_data['Produk'].value_counts()/raw_data.shape[0])
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Ukuran Sebaran pada Data Interval dan Rasio</h2><h3 class="text-cell-section-header">Rentang (<em>range</em>)</h3>
<p>Rentang adalah jarak antara nilai maksimum dengan nilai minimum. Semakin besar jarang antara nilai maksimum dan minimum semakin besar pula sebaran datanya. Secara matematis dapat dituliskan sebagai berikut:</p>
<p>&nbsp;</p>
<div class="MathJax_Display"><span id="MathJax-Element-2-Frame" class="MathJax" tabindex="0"><span id="MathJax-Span-16" class="math"><span id="MathJax-Span-17" class="mrow"><span id="MathJax-Span-18" class="mtext">Range</span><span id="MathJax-Span-19" class="mo">=</span><span id="MathJax-Span-20" class="mo">max</span><span id="MathJax-Span-21" class="mo">(</span><span id="MathJax-Span-22" class="mi">X</span><span id="MathJax-Span-23" class="mo">)</span><span id="MathJax-Span-24" class="mo">−</span><span id="MathJax-Span-25" class="mo">min</span><span id="MathJax-Span-26" class="mo">(</span><span id="MathJax-Span-27" class="mi">X</span><span id="MathJax-Span-28" class="mo">)</span></span></span></span></div>
<p>&nbsp;</p>
<p>Dengan X adalah sampel atau populasi yang sedang diamati.</p>
<p></p></div>

```
# Cari nilai rentang dari kolom 'Pendapatan'
print (raw_data['Pendapatan'].max() - raw_data['Pendapatan'].min())
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Variansi</h2><p>Variansi adalah ukuran sebaran pusat yang diperoleh dengan cara menghitung jarak antara tiap titik data pada sampel atau populasi dengan nilai mean. Secara matematis variansi dirumuskan sebagai berikut:</p>
<p></p>
<p>Untuk menghitung variansi kita dapat menggunakan method&nbsp;<code>.var()</code>&nbsp;dari&nbsp;<code>numpy</code>&nbsp;maupun&nbsp;<code>pandas</code>.</p>
<p></p>
<p>Note: Perhatikan bahwa nilai variansi keduanya berbeda. Hal ini karena secara&nbsp;<em>default</em>&nbsp;pandas menggunakan variansi sampel sedangkan numpy menggunakan variansi populasi. Untuk memperoleh hasil yang sama kita dapat menggunakan parameter&nbsp;<code>ddof=0</code>&nbsp;pada method&nbsp;<code>.var()</code>&nbsp;untuk memperoleh variansi populasi.</p>
<p></p></div>

```
# menghitung variansi umur menggunakan method .var() dari pandas
print (raw_data['Pendapatan'].var())
 
# menghitung variansi umur menggunakan method .var() dari numpy
print (np.var(raw_data['Pendapatan']))

# mengatur variansi populasi dengan method `.var()` dari pandas
print (raw_data['Pendapatan'].var(ddof=0))
```

<div id="theorycontent" style="overflow-x: scroll; display: block;"><h2>Deviasi Baku (Standard Deviation)</h2><p>Deviasi baku adalah ukuran sebaran pusat yang diperoleh dengan cara menarik akar kuadrat dari hasil perhitungan variansi. Hal ini dilakukan karena nilai variansi umumnya memiliki nilai yang lebih besar daripada nilai aslinya sebagai efek dari pengkuadratan dan ini menjadikan variansi sulit untuk diinterpretasikan. Secara matematis deviasi baku dapat dirumuskan sebagi berikut:</p>
<p></p>
<p>Kita dapat menghitung deviasi baku menggunakan method&nbsp;<code>.std</code>&nbsp;dari&nbsp;<code>numpy</code>&nbsp;maupun&nbsp;<code>pandas.</code></p>
<p><code></code></p></div>

```
# menghitung deviasi baku sampel pendapatan menggunakan method std() dari pandas
print (raw_data['Pendapatan'].std())
 
# menghitung deviasi baku sampel pendapatan menggunakan method std() dari numpy
print (np.std(raw_data['Pendapatan'], ddof = 1))
```

Untuk menghitung variansi dan standar deviasi sampel pada kolom 'Pendapatan' dan 'Harga', maka kode yang dapat digunakan adalah:

`raw_data[['Pendapatan', 'Harga']].agg([np.var, np.std], ddof=1)`

<p></p>
<p>Perhatikan bahwa hanya variabel bertipe numerik saja yang dihitung nilai korelasinya. Selain itu secara&nbsp;<em>default</em>&nbsp;<em>method</em>&nbsp;<code>.corr()</code>&nbsp;memiliki parameter&nbsp;<code>method</code>&nbsp;bernilai&nbsp;<code>pearson</code>&nbsp;sehingga nilai korelasi di atas adalah korelasi Pearson. Untuk menggantinya menjadi korelasi Kendall atau korelasi Spearman kita cukup menambahkan parameter&nbsp;<code>method='kendall'</code>&nbsp;atau&nbsp;<code>method='spearman'</code>&nbsp;pada&nbsp;<em>method</em>&nbsp;<code>.corr()</code>&nbsp;sebagaimana contoh berikut:</p>

<p></p>

```
# menghitung korelasi dari setiap pasang variabel pada raw_data
print (raw_data.corr())

# mencari korelasi 'kendall' untuk tiap pasang variabel
print (raw_data.corr(method='kendall'))
 
# mencari korelasi 'spearman' untuk tiap pasang variabel
print (raw_data.corr(method='spearman'))
```
